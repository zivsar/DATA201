## Introduction

For our dataset we used the Thyroid Cancer Risk Dataset found [here](https://www.kaggle.com/datasets/bhargavchirumamilla/thyroid-cancer-risk-dataset) on Kaggle. It was originally derived from [this study](https://doi.org/10.1007/s00405-023-08299-w) by Borzooei et al. It contains observations about a cohort of 383 thyroid cancer patients, each of whom was followed for a minimum of 10 years within a 15 year time period. It tracks thirteen different relevant characteristics of these patients, with the aim of assessing their risk of recurrance.

### Measurements

- **Age**: The age of the patient at diagnosis
- **Gender**: The patient's gender (Male or Female)
- **Country**: The patient's country of origin
- **Ethnicity**: The patient's ethinicty (African, Asian, Caucasian, Hispanic, Middle Eastern)
- **Family_History**: Whether or not the patient had any known family history of thyroid cancer
- **Radiation_Exposure**: Whether the patient had ever been exposed to the kind of radiation that increases the likelihood of thyroid cancers
- **Iodine_Deficiency**: Whether the patient had a prior history of iodine deficiency
- **Smoking**: Whether the patient had ever smoked
- **Obesity**: Whether the patient was obese
- **Diabetes**: Whether the patient had any history of diabetes
- **TSH_Level**: Thyroid stimulating hormone levels (mU/L)
- **T3_Level**: Triiodothyronine levels (ng/dL)
- **T4_Level**: Thyroxine levels (ng/dL)
- **Nodule_Size**: The size in centimeters of the largest nodule (if any) on the thyroid
- **Thyroid_Cancer_Risk**: The risk of the cancer recurring (High, Medium, Low), as calculated based on the American Thyroid Association's [guidelines](https://www.thyroid.org/professionals/calculators/thyroid-cancer-staging-calculator/)
- **Diagnosis**: Whether the cancer is diagnosed as malignant or benign

## Decision Tree Evaluation

```python
# Use scikit-learn DecisionTree Classification or Regression to train, test and predict a categorical or quantitative target variable respectively.
import sklearn as sk

#preprocess data
df = df.replace({'Yes':1, 'No':-1,
                 'Malignant':1, 'Benign':-1,
                 'High':2, 'Medium':1, 'Low':0})
X = df.drop(['Patient_ID', 'Thyroid_Cancer_Risk'], axis=1)
X_one_hot = pd.get_dummies(X, drop_first=True)
y = df['Thyroid_Cancer_Risk']
X_train, X_test, y_train, y_test= sk.model_selection.train_test_split(X_one_hot, y,test_size=0.3, random_state=42, stratify=y, shuffle=True)
```

```python
#create and fit tree
clf = sk.tree.DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(X_train,y_train)

#measure accuracy
y_pred= clf.predict(X_test)
print(f"accuracy score: {np.round(sk.metrics.accuracy_score(y_test, y_pred)*100, 2)}%")

#confusion matrix
cm = sk.metrics.confusion_matrix(y_true=y_test, y_pred=y_pred, normalize='true')
disp = sk.metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low', 'Medium', 'High'])
disp.plot(cmap=plt.cm.YlGnBu)
plt.title('Normalized Confusion Matrix')
plt.figtext(0, 0, ("overall accuracy: " +str(np.round(sk.metrics.accuracy_score(y_test, y_pred)*100, 2))+"%"))
plt.show()
```

The decision tree is good at identifying high risk of reoccurrence, but is likely to confuse low- and medium-risk cases.

```python
#improve predictions using a random forest
rf_cl = sk.ensemble.BaggingClassifier(n_estimators=50, random_state=42)
rf_cl.fit(X_train, y_train)
rf_pred= rf_cl.predict(X_test)

#measure accuracy
print(f"accuracy score: {np.round(sk.metrics.accuracy_score(y_test, rf_pred)*100, 2)}%")

#confusion matrix
cm = sk.metrics.confusion_matrix(y_true=y_test, y_pred=rf_pred, normalize='true')
disp = sk.metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low', 'Medium', 'High'])
disp.plot(cmap=plt.cm.YlGnBu)
plt.title('Normalized Confusion Matrix - Random Forest')
plt.figtext(0, 0, ("overall accuracy: " +str(np.round(sk.metrics.accuracy_score(y_test, rf_pred)*100, 2))+"%"))
plt.show()
```

Using a random forest may increase the accuracy score, but is still not much better at distinguising low- and medium-risk cases.

```python
#improve predictions with xgboost
import xgboost as xgb

xg_cl = xgb.XGBClassifier(objective='binary:logistic', seed=42)
xg_cl.fit(X_train.replace({-1:0}), y_train)
XGy_pred = xg_cl.predict(X_test.replace({-1:0}))

#measure accuracy
print(f"accuracy score: {np.round(sk.metrics.accuracy_score(y_test, XGy_pred)*100, 2)}%")

#confusion matrix
cm = sk.metrics.confusion_matrix(y_true=y_test, y_pred=XGy_pred, normalize='true')
disp = sk.metrics.ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Low', 'Medium', 'High'])
disp.plot(cmap=plt.cm.YlGnBu)
plt.title('Normalized Confusion Matrix')
plt.figtext(0, 0, ("overall accuracy: " +str(np.round(sk.metrics.accuracy_score(y_test, XGy_pred)*100, 2))+"%"))
plt.show()
```

Again, while the accuracy is technically higher, the model completely misses medium-risk cases, misscategorizing them as low-risk.

The original decision tree may be a better tool for catching at least some of the true positives of medium risk for reoccurence so that more people get the care they need, even at the risk of some false positives.

```python
# Plot the tree with tree.plot_tree after the training
class_names = ['Low', 'Medium', 'High']
fig, ax = plt.subplots(figsize=(16, 12))
sk.tree.plot_tree(decision_tree=clf, class_names=class_names, filled=True, feature_names=(X_one_hot.columns), max_depth=10);
```
